```python
import os
import json
import pandas as pd
from collections import Counter

# Define the folder path to search for JSON files
folder_path = "weights_dist_2"

# Initialize an empty list to store the results data frames
results_dfs = []

# Iterate through subfolders
for subdir in os.listdir(folder_path):
    # Initialize lists to store results
    test_accs = []
    train_accs = []
    epochs = []
    learning_rates = []
    optimizers = []
    batch_sizes = []
    num_layers = []
    num_features = []
    num_frames = []
    training_time =[]
    
    # Iterate through JSON files in the subfolder
    for file in os.listdir(os.path.join(folder_path, subdir)):
        # Check if file is a JSON file
        if file.endswith(".json"):
            # Load JSON data
            with open(os.path.join(folder_path, subdir, file), "r") as f:
                data = json.load(f)
            
            # Append values to lists
            test_accs.append(data["best_test_acc"])
            train_accs.append(data["best_train_acc"])
            epochs.append(data["best_epoch"])
            learning_rates.append(data["learning_rate"])
            optimizers.append(data["optimizer"])
            batch_sizes.append(data["batch_size"])
            num_layers.append(data["num_layers"])
            num_features.append(data["num_features"])
            num_frames.append(data["num_frames"])
            training_time.append(data["training_time"])

    # Calculate averages and modes
    avg_test_acc = sum(test_accs) / len(test_accs)
    avg_train_acc = sum(train_accs) / len(train_accs)
    max_train_acc = max(train_accs)
    max_test_acc = max(test_accs)
    avg_epochs = round(sum(epochs) / len(epochs))
    avg_training_time = sum(training_time) / len(training_time)


    # Create a data frame of the results
    results = pd.DataFrame({
        "average_best_test_acc": [avg_test_acc],
        "average_best_train_acc": [avg_train_acc],
        "max_train_acc" : [max_train_acc],
        "max_test_acc": [max_test_acc],
        "average_best_epoch": [avg_epochs],
        "avg_training_time": [avg_training_time]

    })

    results_dfs.append(results)


for i, df in enumerate(results_dfs):
    subdir_name = os.listdir(folder_path)[i]
    print(f"Results for subfolder {subdir_name}:")
    print(df)
    print("\n")
```
